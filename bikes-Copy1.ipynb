```python
def get_closest(vertices, lat, lon):
    verts = pd.DataFrame(vertices, columns=['lon', 'lat'])
    verts['dist'] = verts.apply(lambda x: abs(x['lon'] - lon) + abs(x['lat'] - lat), axis=1)
    return verts.sort_values('dist').head(1)
```
